# Отчет по детекции кликфродов

### Задача:
Выявление кликфродов среди пользователей на основе данных о времени ответа, количестве кликов и других характеристиках.

### Методология решения:
1. **Анализ времени кликов:**
   - Использование графиков рассеяния для изучения среднего времени кликов по различным параметрам, выявление аномалий во времени ответа.

2. **Оценка количества кликов от каждого пользователя:**
   - Определение общего количества кликов и выделение аномальных пользователей, совершивших более 5 кликов.

3. **Оценка времени ответа аномальных пользователей:**
   - Анализ среднего времени ответа аномальных пользователей, выявление пользователей, отвечающих в среднем менее чем за 5 секунд.

4. **Анализ кликов по различным параметрам:**
   - Изучение кликов аномальных пользователей с учетом геолокации, типа устройства, типа браузера и других параметров.

5. **Построение гистограммы кликов для аномальных пользователей:**
   - Визуализация распределения кликов аномальных пользователей с использованием гистограммы.

6. **Обработка категориальных фичей:**
   - Нормализация данных и обработка категориальных фичей для использования в моделях машинного обучения.
7. **Применение K-means для обнаружения аномалий:**
   - Использование K-means с предварительным снижением размерности данных с помощью PCA. Обнаружение и визуализация аномальных пользователей.

8. **Применение DBSCAN для обнаружения аномалий:**
   - Использование DBSCAN для обнаружения и визуализации аномальных пользователей.

9. **Применение Isolation Forest для обнаружения аномалий:**
   - Использование Isolation Forest для выделения аномалий и визуализации результатов на графике рассеяния.

10. **Применение One-Class SVM для обнаружения аномалий:**
   - Использование One-Class SVM для выявления аномалий и визуализации результатов.

11. **Применение LOF для обнаружения аномалий:**
   - Использование Local Outlier Factor (LOF) для выделения аномалий и визуализации результатов.
12. **Анализ временного ряда**
   - Построение графика по дням и часам с использованием скользящего среднего, выделение аномалий и статистический анализ для лучшего понимания активности пользователей на веб-сайте.

### Результаты:
- Выявлены аномальные паттерны во времени кликов и отклике пользователей.
- Обнаружены пользователи, совершившие необычно большое количество кликов.
- Выделены пользователи, отвечающие на клики в среднем менее чем за 5 секунд.
- Проанализированы характеристики аномальных пользователей по геолокации, типу устройства и другим параметрам.
- Применены различные методы машинного обучения (DBSCAN, Isolation Forest, One-Class SVM, LOF) для детекции фродулентных кликов.
- Представлены визуализации результатов для более наглядного понимания аномалий.

### Выводы:
Использование разнообразных методов анализа и машинного обучения позволяет более эффективно выявлять фродулентные клики среди пользователей, предоставляя комплексный подход к анализу данных и принятию решений.

## Анализ группы событий

```python
# Выделение кликов, которые были сделаны после показа (дубликаты по столбцу 'eid')

# Поиск дубликатов по столбцу 'eid'
duplicate_mask = df['eid'].duplicated(keep=False) 

# Выбор строк с дубликатами и сброс индекса
df_view_to_click = df[duplicate_mask].reset_index()  

df_view_to_click
```

```python
# Рассчет времени, прошедшего между показом и кликом для каждого события
df_time_to_click = df_view_to_click.set_index('index').groupby('eid').diff().dropna().reset_index()

# Переименование столбца с временной разницей
df_time_to_click.rename({'t': 't_diff'}, axis=1, inplace=True)

# Вывод индекса и временной разницы
df_time_to_click[['index', 't_diff']]
```

```python
# Объединение фреймов "показ -> клик" с основным фреймом df по индексу
df = df.reset_index().merge(df_time_to_click[['index', 't_diff']], on=['index'], how='left').drop('index', axis=1)

# Заполнение пропущенных значений нулями
df.fillna(0, inplace=True)

# Вывод обновленного DataFrame df
df
```

```python
# Фильтрация DataFrame для получения данных о кликах
df_cliks = df[df['tn'] == 1]

# Вывод статистики по времени между показом и кликом
df_cliks.describe()['t_diff']
```

```python
# Визуализация распределения времени между показом и кликом с обнаружением выбросов

# Создание графиков
fig, ax = plt.subplots(1, 2, figsize=(15, 6), dpi=100)

# Гистограмма и ядерная оценка плотности распределения
sns.histplot(data=df_cliks, x='t_diff', bins=100, ax=ax[0], kde=True)

# Вставной график с уменьшенным диапазоном для лучшего рассмотрения распределения
ins = ax[0].inset_axes([0.4, 0.4, 0.6, 0.6])
sns.histplot(data=df_cliks[df_cliks['t_diff'] < 7], x='t_diff', bins=100, ax=ins, kde=True, color='orange')

# Ящик с усами для визуализации выбросов
sns.boxplot(data=df_cliks, x='t_diff', ax=ax[1], color='orange')

# Вывод информации о выбросах (время между показом и кликом более 200 секунд)
df_cliks[df_cliks['t_diff'] > 200]
```

```python
# Выделение подозрительных кликов: меньше 2 секунд после показа или без показа
suspicious_cliks = df_cliks[(df_cliks['t_diff'] < 2) | (df_cliks['t_diff'] == 0)]
suspicious_cliks
```

```python
# Создание функции для построения круговой диаграммы с учетом порога значений
def pie_by_threshold(df, ax, label_col, threshold, size_col='t_diff', other_name='others'):
    '''
    Функция для построения круговой диаграммы с учетом порога значений.

    Параметры:
    - df: DataFrame с данными
    - ax: объект оси для отображения графика
    - label_col: столбец, по которому строится круговая диаграмма
    - threshold: пороговое значение для отделения данных выше и ниже порога
    - size_col: столбец, используемый для вычисления размеров сегментов
    - other_name: название категории для данных ниже порога

    Возвращает:
    None (отображает круговую диаграмму на переданной оси)
    '''
    
    # Группировка данных и вычисление процентного соотношения
    df_sum = df.groupby(label_col).agg({size_col: 'count'}).apply(lambda x: 100 * x / float(x.sum()))
    df_sum = df_sum.reset_index()
    
    # Разделение данных на две категории: выше и ниже порога
    df1 = df_sum[df_sum[size_col] >= threshold].copy()
    df2 = df_sum[df_sum[size_col] < threshold].copy()
    
    # Создание новой строки для сбора данных ниже порога
    data = {
        label_col: [other_name],
        size_col: [df2[size_col].sum()],
    }
    new_row = pd.DataFrame(data)
    df_last = pd.concat([df1, new_row])
    
    # Создание круговой диаграммы
    sizes = df_last[size_col]
    labels = df_last[label_col]
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140, rotatelabels=90)


# Создание сетки из 3x3 графиков
fig, ax = plt.subplots(3, 3, figsize=(15, 15), dpi=100)

# Определение столбцов для круговых диаграмм и их пороговых значений
columns = {
    'user': 2.2,  # Задание порогового значения для столбца 'user'
    'geo': 5.0,   # Задание порогового значения для столбца 'geo'
    'cnethash': 2.0,  # Задание порогового значения для столбца 'cnethash'
    'devtype': 5.0,   # Задание порогового значения для столбца 'devtype'
    'brotype': 2.0,   # Задание порогового значения для столбца 'brotype'
    'bsize': 5.0,     # Задание порогового значения для столбца 'bsize'
    'siteid': 5.0,    # Задание порогового значения для столбца 'siteid'
    'domhash': 4.0,   # Задание порогового значения для столбца 'domhash'
    'urlhash': 3.0    # Задание порогового значения для столбца 'urlhash'
}

# Итерация по осям для построения круговых диаграмм
i = 0
for row in ax:
    for col in row:
        # Получение текущего столбца для построения круговой диаграммы
        label_col = list(columns.keys())[i]  
        
        # Вызов функции для построения круговой диаграммы с учетом порога значений
        pie_by_threshold(df=suspicious_cliks, ax=col, label_col=label_col, threshold=columns[label_col])
        
        # Установка заголовка для текущего графика
        col.set_title(label_col)  
        i += 1

# Автоматическое распределение графиков внутри фигуры
plt.tight_layout()
```

```python
# Построение графиков среднего времени клика для различных параметров

# Список столбцов для анализа
columns = [
    'user',
    'geo',
    'cnethash',
    'devtype',
    'brotype',
    'bsize',
    'siteid',
    'domhash',
    'urlhash'
]

# Создание сетки из 3x3 графиков
fig, ax = plt.subplots(3, 3, figsize=(15, 15), dpi=100)

# Итерация по осям для построения графиков
i = 0
for row in ax:
    for col in row:
        label_col = columns[i]  # Получение текущего столбца для анализа
        # Группировка данных и вычисление среднего времени клика для каждого значения параметра
        clicks_per_feature = suspicious_cliks.groupby(label_col)['t_diff'].mean().reset_index()
        # Построение точечного графика
        sns.scatterplot(data=clicks_per_feature, x=label_col, y='t_diff', alpha=0.7, ax=col)
        i += 1
```

## Анализ кликов юзеров

```python
# Группировка данных по столбцам 'user' и 'tn', подсчет размера каждой группы
df.groupby(['user', 'tn']).size()
```

```python
# Вычисление количества кликов от каждого пользователя
clicks_per_user = df[df['tn'] == 1].groupby('user').size().reset_index(name='click_count')

# Объединение данных о количестве кликов с основным DataFrame по столбцу 'user'
df = pd.merge(left=df, right=clicks_per_user, on='user', how='left').fillna(0)

# Выделение аномальных пользователей (если количество кликов больше заданного порога 5)
anomalous_users = df[(df['click_count'] > 5) & (df['tn'] == 1)].sort_values('click_count', ascending=False)

# Вывод аномальных пользователей с количеством кликов
anomalous_users[['user', 'click_count']].drop_duplicates('user')
```

```python
# Вычисление общего количества кликов, сделанных аномальными пользователями
anomalous_users['tn'].sum()
```

### Анализ времени кликов аномальных юзеров

```python
# Вычисление среднего времени клика и количества кликов для аномальных пользователей
t_diff_per_anomalous_users = anomalous_users.groupby('user')[['click_count', 't_diff']].mean().reset_index().sort_values('t_diff')

# Вывод DataFrame с информацией о среднем времени клика и количестве кликов для аномальных пользователей
t_diff_per_anomalous_users
```

```python
# Фильтрация аномальных пользователей, среднее время клика которых меньше 5 секунд
t_diff_per_anomalous_users[t_diff_per_anomalous_users['t_diff'] < 5]
```

```python
def anomalous_users_clicks_per_feature(feature, anomalous_users=anomalous_users):
    '''
    Функция для агрегации параметров по аномальным юзерам
    
    Параметры:
    - feature: столбец, по которому осуществляется агрегация
    - anomalous_users: DataFrame с аномальными пользователями
    
    Возвращает:
    DataFrame с агрегированными данными о количестве кликов для каждого значения параметра
    '''
    return anomalous_users.groupby(feature)['tn'].size().reset_index().rename({'tn': 'click_count'}, axis=1).sort_values('click_count', ascending=False)
```

```python
def plot_scatter_with_threshold_y(data, x, y, threshold_y, label, ax):
    '''
    Функция для построения точечной диаграммы с учетом порога значения по оси y.

    Параметры:
    - data: DataFrame с данными
    - x: столбец, отображаемый по оси x
    - y: столбец, отображаемый по оси y
    - threshold_y: пороговое значение по оси y
    - label: название для метки данных выше порога
    - ax: объект оси для отображения графика

    Возвращает:
    None (отображает точечную диаграмму на переданной оси)
    '''
    # Выделение данных выше и ниже порога
    more_than_threshold = data[y].values > threshold_y
    less_than_threshold = data[y].values < threshold_y
    
    # Построение точечной диаграммы для данных ниже порога
    sns.scatterplot(x=data[x].values[less_than_threshold], y=data[y].values[less_than_threshold], ax=ax)
    
    # Построение точечной диаграммы для данных выше порога
    sns.scatterplot(x=data[x].values[more_than_threshold], 
                    y=data[y].values[more_than_threshold], 
                    color='red', 
                    label=label,
                    ax=ax,
                    marker='x')
    
    # Добавление горизонтальной линии на уровне порога
    ax.axhline(y=threshold_y, ls=':', color='red', lw=2)
    
    # Добавление меток для данных выше порога
    for i, txt in enumerate(data[x].values[more_than_threshold]):
        ax.text(
            data[x].values[more_than_threshold][i],
            data[y].values[more_than_threshold][i],
            f'{data[x].values[more_than_threshold][i]}',
            fontsize=8,
            color='black',
            ha='center',
            va='bottom'
        )
```

# Выявление аномалий

## Статистический анализ: обнаружение многомерных выбросов с помощью расстояния Махаланобиса

При работе с многомерными данными простых методов обнаружения одномерных выбросов, таких как Z-показатель или тест Граббса, может быть недостаточно. Вместо этого мы можем использовать расстояние Махаланобиса, статистическую меру, введенную П.К. Махаланобисом в 1936 году. По сути, это многомерный эквивалент Z-показателя. 

---
*Расстояние Махаланобиса учитывает ковариацию между переменными, что помогает выявить закономерности, которые могут быть необнаружимы в отдельных переменных.*

---

Когда данные некоррелированы, расстояние Махаланобиса равно евклидову расстоянию.

Учитывая точку данных x в распределении со средним значением µ и ковариационной матрицей S, расстояние Махаланобиса D вычисляется как:

D² = (x — μ)'S^(-1)(x — μ)

Квадрат расстояния Махаланобиса соответствует распределению хи-квадрат, что позволяет нам выполнить проверку гипотезы для обнаружения выбросов.

### Реализация

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2
from sklearn.covariance import EmpiricalCovariance

def calculate_mahalanobis(data):
    """
    Функция для вычисления расстояния Махаланобиса для заданных данных.

    Parameters:
    - data (array-like): Многомерные данные, для которых нужно вычислить расстояние Махаланобиса.

    Returns:
    - m_dist (numpy.ndarray): Массив расстояний Махаланобиса для каждой точки данных.
    """
    # Инициализация объекта EmpiricalCovariance и оценка ковариационной матрицы
    robust_cov = EmpiricalCovariance().fit(data)
    
    # Вычисление расстояний Махаланобиса для каждой точки данных
    m_dist = robust_cov.mahalanobis(data)
    
    return m_dist

# Вычисление расстояний Махаланобиса для данных click_data_scaled
m_dist = calculate_mahalanobis(click_data_scaled)

# Определение уровня значимости для обнаружения аномалий
alpha = 0.05

# Вычисление порога (на основе хи-квадрат распределения)
thresh = chi2.ppf(1 - alpha, df=click_data_scaled.shape[1])

# Создание DataFrame для хранения результатов и копирование данных из click_data_scaled
stats_df = click_data_scaled.copy()

# Добавление расстояний Махаланобиса и меток аномалий в DataFrame
stats_df['mahalanobis_distance'] = m_dist
stats_df['is_anomaly'] = (stats_df['mahalanobis_distance'] > thresh) * 1

# Формирование DataFrame с аномалиями
stats_frauds = stats_df[stats_df['is_anomaly']==True]

# Вывод результатов
stats_frauds
```

### Визуализация

```python
def plot_stats_scatter(x, y):
    """
    Строит график рассеяния для двух выбранных переменных с выделением выбросов.

    Parameters:
    - x (str): Название переменной для оси x.
    - y (str): Название переменной для оси y.

    Returns:
    - None
    """
    # Создание графика рассеяния
    plt.scatter(stats_df[x], 
                stats_df[y], 
                c=~stats_df['is_anomaly'], 
                cmap='viridis',
                marker='o', 
                edgecolors='k',
                label='Inliers')
    
    # Выделение аномалий красными крестами
    plt.scatter(stats_df[stats_df['is_anomaly'].astype(bool)][x], 
                stats_df[stats_df['is_anomaly'].astype(bool)][y], 
                c='red', marker='x', label='Anomalies')
    
    # Добавление меток осей и заголовка
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title('Scatter Plot of Mahalanobis Distance Anomalies')
    
    # Добавление легенды
    plt.legend()
    
# Пример использования функции
plot_stats_scatter(x='t_diff', y='click_impression_ratio')
```

## PCA

Перед применением применением различных методов машиннного обучения для выявления аномалий в многомерных данных, мы стремимся учесть сложные взаимосвязи между переменными и эффективно кластеризовать данные с учетом снижения размерности. Для этого воспользуемся PCA для создания нового пространства признаков, где алгоритмы смогут более точно выделить аномалии

---
*Применение PCA предоставляет возможность учесть ковариацию между переменными и выделить наиболее важные направления изменчивости, что существенно повышает эффективность обнаружения аномалий.*

---

```python
from sklearn.decomposition import PCA

# Инициализация объекта PCA с указанием количества компонент (в данном случае, 2)
pca = PCA(n_components=2, random_state=42)

# Применение PCA к нормализованным данным click_data_scaled
X_pca = pca.fit_transform(click_data_scaled)

# Построение scatter plot для визуализации данных в новом пространстве признаков
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', alpha=0.5)
plt.title('Scatter plot of the data before K-Means')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2');
```

## K-means

При работе с K-means для выявления аномалий применяется подход, аналогичный расстоянию Махаланобиса, но с учетом центров кластеров. В отличие от простых методов обнаружения одномерных выбросов, K-means позволяет учесть структуру данных в многомерном пространстве.

---
*Когда данные кластеризуются с использованием K-means, каждая точка данных присваивается к одному из кластеров, а центр этого кластера становится представителем этой группы. Для оценки аномалий можно использовать расстояние от каждой точки данных до центра её соответствующего кластера.*

---

Допустим, у нас есть точка данных x, принадлежащая кластеру с центром μ. Тогда расстояние от x до центра кластера D вычисляется как:

D = ||x - μ||

Таким образом, мы оцениваем аномалии на основе того, насколько точки данных отклоняются от центра своего кластера. При наличии аномалий точки данных могут оказаться далеко от центров кластеров, что позволяет идентифицировать такие аномалии в данных, используя K-means.

### Реализация

```python
from sklearn.cluster import KMeans

# Инициализация модели K-means с указанием количества кластеров (в данном случае, 2)
kmeans = KMeans(n_clusters=2, random_state=42)

# Обучение модели на данных после применения PCA
kmeans.fit(X_pca)

# Прогнозирование меток кластеров: labels
labels = kmeans.predict(X_pca)

# Вычисление расстояний до центров кластеров
distances = kmeans.transform(X_pca)

# Вычисление аномальных юзеровго балла (расстояние до ближайшего центроида)
scores = distances.max(axis=1)

# Нормализация оценки до диапазона [0, 1] для цветового отображения
normalized_scores = (scores - min(scores)) / (max(scores) - min(scores))

# Создание объекта Series с нормализованными аномальными баллами
normalized_scores_series = pd.Series(normalized_scores, name='anomaly_score')

# Создание DataFrame для хранения результатов и копирование данных из click_data_scaled
kmeans_df = click_data_scaled.copy()

# Соединение исходного DataFrame с нормализованными баллами
data_with_scores = pd.concat([kmeans_df, normalized_scores_series], axis=1)

# Выбор порога на основе 95-го процентиля
threshold = np.percentile(normalized_scores_series, 95)

# Выделение аномальных наблюдений, превышающих установленный порог
kmeans_frauds = data_with_scores[data_with_scores['anomaly_score'] > threshold]

# Вывод DataFrame kmeans_frauds с аномальными наблюдениями
kmeans_frauds
```

### Визуализация

```python
# Создание scatter plot с учетом меток кластеров после применения K-Means
plt.figure(figsize=(10, 7))

# Разноцветные точки для каждого кластера
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', alpha=0.5)

# Выделение красных точек для центроидов кластеров
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')

# Настройка заголовка и осей графика
plt.title('Scatter plot данных после применения K-Means')
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
```

- Использование 95-го процентиля обычно связано с исключением 5% самых далеких точек, что уменьшает влияние выбросов на расчет порогового значения. Такой подход улучшает устойчивость оценки и делает ее менее чувствительной к крайним значениям.

## DBSCAN

Метод, который не просто разбивает данные на кластеры, но учитывает пространственную плотность точек данных — DBSCAN (Пространственная кластеризация приложений с шумом на основе плотности).  
Сила DBSCAN заключается в его способности идентифицировать кластеры различных форм и размеров, а также в его устойчивости к шуму и выбросам, что делает его мощным инструментом для обнаружения аномалий.  
Концепция DBSCAN (пространственная кластеризация приложений с шумом на основе плотности) заключена в самом названии — это алгоритм кластеризации на основе плотности. В отличие от K-Means, который разбивает данные на сферические кластеры и может ошибочно интерпретировать выбросы как небольшие кластеры, DBSCAN находит области с высокой плотностью и отделяет их от областей с низкой плотностью.

---
Эта уникальная характеристика позволяет ему обнаруживать кластеры произвольной формы и размера, а также обрабатывать шум и выбросы, что делает его особенно подходящим для обнаружения аномалий.

---


### Реализация

```python
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
'''
Выбор оптимального значения параметра eps (радиус окрестности) в алгоритме DBSCAN 
(Density-Based Spatial Clustering of Applications with Noise). 

Алгоритм DBSCAN определяет кластеры в данных, учитывая их плотность. 
Параметр eps задает радиус окрестности, в пределах которого точки считаются соседями.

Прежде чем выбрать значение eps, мы вычисляем расстояния до k-го ближайшего соседа для каждой точки данных. 
Затем строится график этих расстояний. На графике мы ищем точку, где происходит "изгиб" (inflexion point),
что может свидетельствовать о переходе от плотного кластера к разреженному пространству. 
Выбранное методом локтя расстояние в этой точке используется в качестве значения eps для алгоритма DBSCAN. 
В данном коде chosen_eps представляет собой выбранное значение eps.
'''

# Рассчет расстояний до k-го ближайшего соседа
k = 3
neigh = NearestNeighbors(n_neighbors=k)
neigh.fit(click_data_scaled)
distances, _ = neigh.kneighbors(click_data_scaled)

# График k-distance
distances = np.sort(distances[:, -1], axis=0)
plt.plot(distances)
plt.grid(True)
plt.xlabel('Номер точки данных')
plt.xticks(range(0, 1547, 150))
plt.ylabel(f'{k}-е расстояние')
plt.show()

# Выбор эпсилон (радиус окрестности) по графику k-distance
chosen_eps = distances[750]
chosen_eps
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

'''
В данном коде используется алгоритм DBSCAN с разными значениями параметра min_samples,
чтобы оценить, как изменение этого параметра влияет на количество образующихся кластеров. 
min_samples определяет минимальное количество точек, которые должны находиться в радиусе eps, 
чтобы образовать кластер.

На графике отображается зависимость количества кластеров от значения min_samples. 
Это позволяет выбрать оптимальное значение min_samples в контексте конкретной задачи или набора данных. 
Оптимальное значение min_samples будет зависеть от структуры данных и требований к кластеризации.
'''


# Используем разные значения min_samples и смотрим, сколько кластеров образуется
min_samples_values = range(1, 20)
num_clusters = []

# Проходим по разным значениям min_samples
for min_samples_val in min_samples_values:
    # Создаем объект DBSCAN с выбранным eps и min_samples
    dbscan = DBSCAN(eps=chosen_eps, min_samples=min_samples_val)
    
    # Применяем алгоритм DBSCAN к отмасштабированным данным
    labels = dbscan.fit_predict(click_data_scaled)
    
    # Вычисляем количество кластеров, учитывая, что -1 обозначает выбросы (аномалии)
    num_clusters.append(len(set(labels)) - (1 if -1 in labels else 0))

# График количества кластеров от min_samples
plt.plot(min_samples_values, num_clusters, marker='o')
plt.xlabel('min_samples')
plt.grid(True)
plt.ylabel('Количество кластеров')
```

```python
# Определение модели DBSCAN с выбранным параметром расстояния (eps) и минимальным количеством соседей (min_samples)
dbscan = DBSCAN(eps=chosen_eps, min_samples=10)

# Обучение модели на данных и прогноз меток кластеров
labels = dbscan.fit_predict(X_pca)

# Определение ядерных (core) и аномальных точек
core_samples_mask = np.zeros_like(dbscan.labels_, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
labels = dbscan.labels_

# Получение количества кластеров (без учета выбросов) и количества точек-выбросов
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print(f'Оценка количества кластеров: {n_clusters_}')
print(f'Оценка количества точек-выбросов: {n_noise_}')

# Пометка и отображение аномалий
anomaly_mask = labels == -1
anomalies = X_pca[anomaly_mask]

dbscan_frauds = click_data_scaled[anomaly_mask]
dbscan_frauds
```

### Визуализация

```python
# Построение данных, закрашивая их в соответствии с кластерами DBSCAN
plt.figure(figsize=(10,7))
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]  # Черный цвет для выбросов
    class_member_mask = (labels == k)

    # Отображение точек-ядерных и выбросов с разными размерами маркеров
    xy = X_pca[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=10)

    xy = X_pca[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=5)

plt.plot(anomalies[:, 0], anomalies[:, 1], 'x', markerfacecolor='red', markeredgecolor='red', markersize=5, label='Аномалии')
plt.title(f'Оценка количества кластеров: {n_clusters_}')
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
plt.legend();
```

### Вывод

Полученные результаты:
1. Оценка количества кластеров: 9
2. Оценка количества точек-выбросов: 105
- Означает, что алгоритм нашел 9 различных групп точек данных, где точки каждой группы плотно упакованы вместе в пространстве признаков.
Примерное количество точек шума: 105;
- DBSCAN пометил 105 точку данных как шум. Это точки, которые не принадлежат ни одному кластеру, поскольку не являются частью плотной области.

В DBSCAN точки шума могут быть либо аномалиями/выбросами, либо точками, расположенными между кластерами. В этом контексте эти точки шума потенциально можно рассматривать как аномалии или выбросы .

## Isolation forest

Изоляционный лес (Isolation Forest) представляет собой метод обнаружения аномалий, который основывается на идее изоляции аномалий в пространстве признаков. Этот метод не только разбивает данные на кластеры, но и определяет степень изоляции каждой точки данных от остальных, что делает его эффективным инструментом для выявления аномалий.

---
Isolation Forest выделяется среди методов аномалий за его способность быстро и эффективно выделять аномальные точки. Этот алгоритм строит деревья решений, разбивая пространство признаков так, чтобы аномалии оказывались изолированными от основной массы данных.

---

Ключевые особенности Isolation Forest:

1. Изоляция аномалий: Путем построения деревьев решений метод случайным образом разбивает пространство признаков, изолируя аномалии от основной массы данных.
2. Глубина деревьев: Чем быстрее точка данных будет изолирована, тем более вероятно, что она является аномальной. Глубина деревьев служит мерой аномальности точек данных.
3. Агрегация результатов: Для каждой точки данных вычисляется средняя глубина изоляции по всем деревьям. Точки с более низкой средней глубиной считаются более аномальными.


### Реализация

```python
# Импорт модели Isolation Forest из библиотеки sklearn.ensemble
from sklearn.ensemble import IsolationForest

# Определение модели Isolation Forest с заданным параметром contamination (доля аномалий)
isolation_forest = IsolationForest(contamination=0.05, random_state=42)

# Обучение модели и прогнозирование аномалий
anomaly_mask = isolation_forest.fit_predict(X_pca)

# Создание копии DataFrame с исходными данными
anomalies = click_data_scaled.copy()

# Добавление столбца 'is_anomaly', где 1 - аномалия, 0 - не аномалия, на основе меток из Isolation Forest
anomalies['is_anomaly'] = [1 if x == -1 else 0 for x in anomaly_mask]

# Выделение данных с аномалиями на основе добавленной метки 'is_anomaly'
iso_forest_frauds = anomalies[anomalies['is_anomaly'] == 1]

iso_forest_frauds
```

### Визуализация

```python
# Создание DataFrame с исходными данными и метками аномалий
iso_forest_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
iso_forest_df['Isolation Forest Anomaly'] = anomaly_mask

# Построение графика, цвета кодируются метками аномалий
plt.figure(figsize=(10, 7))
colors = ['blue' if label == 1 else 'red' for label in iso_forest_df['Isolation Forest Anomaly']]
plt.scatter(iso_forest_df['Principal Component 1'], iso_forest_df['Principal Component 2'], c=colors, s=10)

plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

### Вывод
По результатам использования модели Isolation Forest с параметром contamination (доля аномалий) равным 5%, было обнаружено 78 аномалий в наборе данных. Эта модель идентифицирует аномалии на основе того, насколько отличаются точки данных от общего паттерна.

Для оценки аномальности каждой точки данных была использована метка из Isolation Forest: -1, если точка считается аномальной, и 0, если не аномальной. Далее был создан новый DataFrame iso_forest_frauds, включающий только те данные, которые модель отметила как аномалии (с меткой 1).

Такой подход позволяет выделить и проанализировать конкретные случаи, которые могут отличаться от общего тренда в данных.

## One class SVM

Метод One-Class SVM (Support Vector Machine) — это алгоритм обнаружения аномалий, который строит модель, исходя из представления о "нормальных" данных, и затем идентифицирует аномалии как отклонения от этого представления.

---
One-Class SVM представляет собой мощный инструмент в обнаружении аномалий, опираясь на идею создания модели для "нормальных" данных и выявления отклонений от этого стандарта.

---

Ключевые характеристики One-Class SVM:

1. Моделирование "нормальных" данных: Алгоритм строит гиперплоскость, которая наилучшим образом описывает "нормальные" данные, опираясь на метод опорных векторов.
2. Отделение аномалий: Точки данных, находящиеся за пределами границы, установленной вокруг "нормальных" данных, считаются аномалиями.
3. Контроль ширины границы: Ширина границы, определяющей, что считать аномальным, регулируется параметром nu (нормально ошибка).
4. Адаптивность к структуре данных: One-Class SVM обеспечивает адаптивность к сложным формам данных, что позволяет выявлять аномалии в различных сценариях.

Применение One-Class SVM особенно полезно, когда "нормальные" данные преобладают, и аномалии представляют собой отклонения от этой нормы. 

### Реализация

```python
from sklearn.svm import OneClassSVM

# Инициализация модели One-Class SVM с указанием параметров
svc = OneClassSVM(nu=0.1, kernel="rbf", gamma=0.1)

# Обучение модели на данных
svc.fit(X_pca)

# Прогнозирование меток для всего набора данных
svc_anomalies = np.array([1 if x == -1 else 0 for x in svc.predict(X_pca)])

# Создание DataFrame с исходными данными и метками аномалий от One-Class SVM
svc_frauds = pd.concat([click_data_scaled, pd.DataFrame({"is_anomaly": svc_anomalies})], axis=1)

# Выделение данных с аномалиями на основе добавленной метки 'is_anomaly'
svc_frauds = svc_frauds[svc_frauds['is_anomaly'] == 1]

svc_frauds
```

### Визуализация

```python
# Генерация сетки для построения графика
xx, yy = np.meshgrid(np.linspace(X_pca[:, 0].min(), X_pca[:, 0].max(), 500),
                     np.linspace(X_pca[:, 1].min(), X_pca[:, 1].max(), 500))

# Прогнозирование оценок по сетке
Z = svc.decision_function(np.c_[xx.ravel(), yy.ravel()])

# Изменение формы оценок для сетки
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 10))

# Построение границы решения
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.PuBu)
a = plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='darkred')

# Построение точек данных
b1 = plt.scatter(X_pca[svc_anomalies == 0, 0], X_pca[svc_anomalies == 0, 1], c='white', s=20, edgecolors='k')
b2 = plt.scatter(X_pca[svc_anomalies == 1, 0], X_pca[svc_anomalies == 1, 1], c='red', s=20, edgecolors='k')

plt.axis('tight')
plt.xlim((X_pca[:, 0].min(), X_pca[:, 0].max()))
plt.ylim((X_pca[:, 1].min(), X_pca[:, 1].max()))

plt.legend([a.collections[0], b1, b2],
           ["learned frontier", "normal points", "anomalous points"],
           loc="best")

plt.title("One-Class SVM Anomaly Detection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
```

### Вывод

Модель One-Class SVM, примененная в данном контексте с параметрами nu=0.1, kernel="rbf", и gamma=0.1, классифицирует точки данных как аномалии на основе своей способности выделять область, содержащую нормальные данные. В данной ситуации 154 точки были отмечены как аномалии. Это объясняется следующим образом:

1. **Обучение на нормальных данных:**
   - Модель обучена на данных, представляющих собой в основном нормальные случаи. One-Class SVM строит гиперплоскость, которая охватывает область в пространстве признаков, считаемую "нормальной".

2. **Гиперплоскость и "окно":**
   - Формируется гиперплоскость, которая определяет "окно" вокруг нормальных данных. Точки, расположенные вне этого "окна", тем более если они значительно удалены, могут считаться аномальными.

3. **Метки аномалий:**
   - При прогнозировании меток для всего набора данных, One-Class SVM отмечает точки, которые находятся за пределами определенной границы, как аномалии (метка -1).

4. **Контроль параметра nu:**
   - Параметр `nu` устанавливает верхнюю границу доли данных, которые модель считает аномальными. В данном случае, с установкой `nu=0.1`, примерно 10% данных считаются аномалиями.

Таким образом, точки, классифицированные как аномалии, выделяются моделью One-Class SVM на основе их значительного отклонения от области, содержащей нормальные данные, формируемой в процессе обучения.

## LOF

Локальный фактор выбросов (LOF) представляет собой алгоритм обнаружения аномалий, который оценивает относительную плотность точек данных в их соседстве. Отличительной чертой LOF является его способность выделять аномалии на основе локальных плотностей, что делает его эффективным инструментом в различных сценариях обнаружения аномалий.

---
*LOF, или локальный фактор выбросов, выделяется своей способностью оценивать локальные плотности точек данных, что делает его эффективным методом обнаружения аномалий в многих сценариях.*

---

Ключевые особенности LOF:

1. **Локальная оценка плотности:** Алгоритм оценивает локальную плотность каждой точки данных, исходя из плотности их соседей.

2. **LOF-значение:** LOF для каждой точки представляет собой отношение локальной плотности точки к локальным плотностям её соседей. Точки с LOF значением выше 1 считаются аномалиями.

3. **Устойчивость к изменению плотности:** LOF адаптивен к изменениям плотности в данных, что позволяет ему выявлять аномалии в кластерах различной плотности.

4. **Беспристрастность к форме кластера:** LOF не зависит от формы кластеров, что делает его подходящим для данных с различными структурами.

LOF особенно эффективен в тех случаях, когда аномалии имеют различные плотности в зависимости от их окружения. Применяется, например, в области мониторинга транзакций, выявления нештатных событий в сетях или обнаружения аномалий в медицинских данных.

### Реализация

```python
# Импорт необходимого класса для обнаружения выбросов
from sklearn.neighbors import LocalOutlierFactor

# Инициализация модели Local Outlier Factor (LOF) с параметрами: количество соседей - 10, уровень загрязнения - 0.1
clf = LocalOutlierFactor(n_neighbors=10, contamination=0.1)

# Прогноз меток аномалий на основе данных после применения PCA
y_pred = clf.fit_predict(X_pca)

# Создание DataFrame с исходными данными и метками аномалий
lof_fraud = pd.concat([click_data_scaled, pd.DataFrame({'is_anomaly': [1 if x == -1 else 0 for x in y_pred]})], axis=1)

# Выбор только тех наблюдений, для которых установлена метка аномалии
lof_fraud = lof_fraud[lof_fraud['is_anomaly'] == 1]

lof_fraud
```

### Визуализация

```python
# Generate colors based on the prediction
colors = np.array(['#377eb8', '#ff7f00'])
plt.scatter(X_pca[:, 0],X_pca[:, 1], c=y_pred, cmap='viridis', alpha=0.5)

plt.title("Local Outlier Factor (LOF)")
pd.Series(y_pred).value_counts()
```

### Вывод

Модель Local Outlier Factor (LOF), примененная в данном контексте с параметрами n_neighbors=10 и contamination=0.1, идентифицирует аномалии на основе изоляции точек данных от своих ближайших соседей. В данной ситуации 154 точки были отмечены как аномалии, и это объясняется следующим образом:

1. **Изоляция от ближайших соседей:**
   - Модель LOF оценивает аномалии, основываясь на том, насколько изолированными они являются от своих ближайших соседей. Установка `n_neighbors=10` указывает, что для каждой точки рассматриваются 10 ближайших соседей.

2. **Уровень загрязнения (contamination):**
   - Параметр `contamination=0.1` определяет долю данных, которые следует считать аномальными. Примерно 10% данных считаются аномалиями в данном случае.

3. **Прогноз меток аномалий:**
   - Метки аномалий прогнозируются моделью LOF при помощи метода `fit_predict(X_pca)`. Точки, для которых прогнозируется метка -1, считаются аномалиями, в то время как метка 0 присваивается нормальным точкам.

4. **Выбор только аномальных точек:**
   - Для создания DataFrame `lof_fraud` были выбраны только те точки, для которых метка аномалии равна 1. Таким образом, этот DataFrame представляет собой подмножество данных, где сосредоточены те точки, которые модель LOF считает аномальными.

В итоге, LOF выделяет аномалии на основе их отклонения от нормальных ближайших соседей, что позволяет выявить потенциально необычные или выделенные точки в наборе данных.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

'''
Этот код предназначен для анализа временного за весь период по дням и по часам о кликах во времени. Он строит два графика:
1. График кликов в день с использованием библиотек matplotlib и seaborn.
2. График кликов в час с использованием скользящего среднего для сглаживания данных.
'''

# Анализ кликов по времени
fig, ax1 = plt.subplots(2, 1, figsize=(30, 15), gridspec_kw={'height_ratios': [1, 2]})

# Анализ кликов по дням
clicks_per_day = df[df['tn'] == 1].set_index('t').resample('D').count().reset_index()
ax1[0].set_ylabel('Количество кликов в день')
sns.lineplot(x=clicks_per_day['t'], y=clicks_per_day['eid'], data=clicks_per_day, ax=ax1[0], marker='o')

# Анализ кликов по часам и скользящее среднее
clicks_per_hour = df[df['tn'] == 1].set_index('t').resample('10min').count()

# Вычисление стандартного отклонения
std_deviation = clicks_per_hour['tn'].std()

# Выделение аномалий
anomalies = clicks_per_hour[np.abs(clicks_per_hour['tn'] - clicks_per_hour['tn'].rolling(window=10).mean()) > 2 * std_deviation]

# Построение графика и скользящего среднего
ax1[1].set_ylabel('Количество кликов в час')
sns.lineplot(x=clicks_per_hour.index, y=clicks_per_hour['eid'], data=clicks_per_hour, ax=ax1[1], label='Исходные данные')
sns.lineplot(x=clicks_per_hour.index, y=clicks_per_hour['tn'].rolling(window=10).mean(), label='Скользящее среднее (10 часов)', ax=ax1[1])
ax1[1].scatter(anomalies.index, anomalies['eid'], color='red', label='Аномалии')

# Установка интервала меток оси x на часовой интервал
ax1[1].xaxis.set_major_locator(mdates.HourLocator(interval=2))

# Форматирование меток оси x в виде часов и минут
ax1[1].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.xticks(rotation=90)
plt.tight_layout()

# Установка пределов оси x в соответствии с данными
ax1[1].set_xlim(clicks_per_hour.index.min(), clicks_per_hour.index.max())

# Добавление легенды
ax1[1].legend()

# Установка подписей и заголовка
ax1[1].set_xlabel('Время')
ax1[1].set_ylabel('Количество кликов')
ax1[1].set_title('График и скользящее среднее')

# Отображение графика
plt.tight_layout()
```
